```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
data = pd.read_csv("UNSW_NB15_training-set.csv")
test = pd.read_csv("UNSW_NB15_testing-set.csv")
```

```python
col = data.select_dtypes(include =["object","category"]).columns
print(col)
```

```python
le = LabelEncoder()
for i in col:
    data[i] = le.fit_transform(data[i])
    test[i] =le.fit_transform(test[i])

   
```

```python
data.head(5)
```

```python
ncol = data.select_dtypes(include =["int64","float64"]).columns
tstncol = test.select_dtypes(include =["int64","float64"]).columns
print(ncol[0])
print(data.dtypes)
```

```python
sc = StandardScaler()
proc = ColumnTransformer(
transformers =[("ncol", StandardScaler(),ncol),
              ])

proc.fit_transform(data)
proct = ColumnTransformer(
transformers =[("tncol", StandardScaler(),tstncol),
              ])
proct.fit_transform(test)
    
```

```python
data.to_csv("preprocessed.csv", index = False)
test.to_csv("preprocessedtest.csv", index = False)
```
